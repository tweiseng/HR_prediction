```python
def sal_class(x):
    if x == "low":
        return 1
    elif x == "medium":
        return 2
    elif x == "high":
        return 3
    
df['salary_class'] = df['salary'].apply(sal_class)

df.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Establishing the base dataset
X = df[['satisfaction_level','last_evaluation','number_project',
        'average_monthly_hours','exp_in_company','work_accident',
        'promotion_last_5years','salary_class']]
y = df['left'] 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
# print(train_test_split(X, y, test_size=0.3, random_state=10))
```

```python
from sklearn.metrics import accuracy_score

logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)

predictions = logmodel.predict(X_test)

# accuracy_score(y_test,predictions)
```

```python
accuracy_score(y_test,predictions)
```

```python
predict = pd.read_csv('hr_predict.csv')

X_predict = predict[['satisfaction_level','last_evaluation','number_project',
                     'average_monthly_hours','exp_in_company','work_accident',
                     'promotion_last_5years','salary_class']]

predictions = logmodel.predict(X_predict)

X_predict = predict

# print(predictions)
X_predict.assign(left_predict = predictions)
```
